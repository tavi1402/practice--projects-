```python
import random
logo = ''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    '''

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']




print("welcome to the Hangman game ",logo)
word_list=["sanna","jyoti","kiran"]

#chose the random word from given word list by using coice function from random module 
chosen_word=random.choice(word_list)
print(f"the computer chose the letter {chosen_word}")

end_of_game=False
lives=6     #to stop the game we are using a number means after 6 wrong many attempts the game will stop.

        
# to print a list of blank "-" space of the same length of chosen word

list=[] 
word_length=len(chosen_word)
for _ in range(word_length):
    list+="_"
print(list)


#iteration starts 

while not end_of_game:
    guess= input("guess the letter ").lower() 
    
    if guess in list:
        print(f"you already guessed this letter before : {guess}")

    for position in range(word_length):
        letter=chosen_word[position]   #(indexing to get to know the poition of lettr in chosen word)
        #print(f"current position : {position}\n current letter : {letter})\n gussed letter : {guess}")
        
        if letter==guess:
            list[position]=letter
    print(list) 
    
    if guess not in chosen_word:
        print(f"you guessed this particulat letter that's not in the word.you loose a life")
        lives -=1
        if lives==0:
            end_of_game=True
            print("you loose!!")
        
    
    if "_" not in list:
        end_of_game=True
        print("you win!!")
   
    print(stages[lives])
    
```
