```python
guess_max_age = int(input("enter the maximum age you are expecting to live "))
current_age=int(input("enter your current age "))
year_remaining= guess_max_age - current_age
days_remaining= year_remaining*365
months_remaining= year_remaining*12
weeks_remaining=year_remaining*52


f"you have {days_remaining} days,{weeks_remaining} weeks, and {months_remaining} months"
```
