# calculating average height with for loop 

```python
# calculating average height with for loop 

student_heights=input("enter the heights of the students in class").split()
for i in range(0, len(student_heights)):
    student_heights[i]= int(student_heights[i])
    
total_height=0
for j in student_heights:
    total_height=total_height+j
print(f"the sum of total height of students is{total_height}") 

total_number_of_students =0
for k in student_heights:
    total_number_of_students+=1
print(f"total numbet of students is {total_number_of_students}")  

average_height= round(total_height/total_number_of_students)
print(average_height)
```

# calculating maximum or heighest with for loop 

```python
student_scores = input("Input a list of student heights ").split()
for n in range(0, len(student_scores)):
  student_scores[n] = int(student_scores[n])

heighest_score=0
for score in student_scores:
    if score > heighest_score:
        heighest_score=score
print(f"the heighest score is {heighest_score}")    
```
